```python
# import urllib library
import requests
import json
import pandas as pd
```

```python
response = requests.get("https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow").text
```

```python
response_info = json.loads(response)
```

```python
response_info
```

```python
type(response_info)
```

```python
response_info['items']
```

```python
normalized = pd.json_normalize(response_info['items'])

df2 = pd.DataFrame.from_dict(normalized)
```

```python
df2
```

```python
df2.columns
```

```python
df2.groupby(['is_answered']).count()
```

```python
df2.groupby('is_answered').size()
```

```python
df2.groupby('owner.user_id')['view_count'].value_counts()
```

```python
by_year = df2.sort_values('view_count')
print('ID con menor numero de vistas: ')
by_year.head(1)
```

```python
df2.groupby('owner.user_id')['owner.reputation'].value_counts()
```

```python
rep = df2.sort_values('owner.reputation', ascending=False)
print('Owner con mejor reputacion: ')
rep.head(1)
```

```python
SELECT tabla_de_aerolineas.NOMBRE_AEROLINEA, tabla_de_aeropuertos.NOMBRE_AEROLINEA, tabla_de_movimientos.DESCRIPCION
FROM tabla_de_vuelos
INNER JOIN tabla_de_vuelos 
ON tabla_de_aerolineas.ID_AEROLINEA = ID_AEROLINEA 
AND tabla_de_aeropuertos.ID_AEROPUETO = ID_AEROPUERTO
AND tabla. demovimientos.ID_MOVIMIENTO = ID_MOVIMIENTO;
```

```python
1. Nombre del aeropuerto con mayor movimientos

SELECT ID_AEROPUERTO
FROM tabla de vuelos
GROUP BY ID_AEROPUERTO
WHERE COUNT(ID_AEROPUERTO)>1;

```

```python
2. Nombre de aerolinea con mayores vuelos

SELECT ID_AEROLINEA
FROM tabla de vuelos
GROUP BY ID_MOVIMIENTO
WHERE COUNT(ID_MOVIMIENTO)>1;
```

```python
3. Dia con mayores vuelos

SELECT DIA, SUM(DIA) AS Vuelos_totales
FROM tabla de vuelos
GROUP BY ID_MOVIMIENTO
ORDER BY ID_MOVIMIENTO;
```

```python
4. Aerolineas con mas de dos vuelos por dia

SELECT ID_AEROLIENA
FROM tabla de vuelos
GROUP BY DIA
WHERE COUNT(DIA)>2;
```
